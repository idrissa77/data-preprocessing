```python
from os import listdir
```

```python
noms_fichiers = listdir("/content/drive/MyDrive/Speeches")
noms_fichiers
/content/Cleaned_speeches_J_A_DATATECH
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(index=range(len(noms_fichiers)), columns = ['Nom_du_fichier', 'Contenu'])
df.head()
```

```python
for i,f_name in enumerate(noms_fichiers):
    f = open("/content/drive/MyDrive/Speeches/" + f_name, "r", encoding="utf-8")
    f_content = f.readlines()
    f.close()
    
    df.at[i, "Nom_du_fichier"] = f_name
    df.at[i, "Contenu"] = f_content[0]
```

```python
df
```

```python
mois = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
mois
```

```python
def SeparerVille(v):
  ville = ''
  for m in mois: 
    if (m in v):
      ville = v[:v.find(m)]
  return ville
```

```python
# Application de la fonction SeparerVille()
df['Ville'] = df.Nom_du_fichier.apply(SeparerVille)
df.head()
```

```python
df.tail()
```

```python
def SeparerDate(c):
  date = c.Nom_du_fichier[len(c.Ville):c.Nom_du_fichier.find(".txt")]
  return date
```

```python
df["Date"] = df.apply(SeparerDate, axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%b%d_%y")
```

```python
df.head()
```

```python
df.drop("Nom_du_fichier", axis = 1, inplace=True)
```

```python
df
```

```python
df["Annee"] = df.Date.dt.year
```

```python
df["Mois"] = df.Date.dt.month
```

```python
df["Jour_ouvre"] = df.Date.dt.day # Jour_ouvre
```

```python
df["Jour_off"] = df.Date.dt.weekday # Weekend
```

```python
df.head()
```

```python
import numpy as np
```

```python
df["weekend"] = np.where((df.Jour_off == 6) | (df.Jour_off == 7), "Oui","Non")
```

```python
df.to_csv('Cleaned_speeches_J_A_DATATECH.csv', index= False)
```
